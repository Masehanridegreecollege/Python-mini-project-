```python
#Read the CSV file into a pandas as pd
import pandas as pd
```

```python
datasetpath='/content/retail_transactions.csv'
```

```python
df=pd.read_csv(datasetpath)
```

```python
df
```

```python
 #Find the total revenue generated from all transactions
 df['total_amount']=df['Quantity']*df['Price']
 total_revenue=df['total_amount'].sum()
 print("Total revenue: ₹",total_revenue)

```

```python


#Identify the most frequently purchased product
most_frequent_purchased_product=df['ProductCategory'].mode()[0]
print("Most frequently purchased ProductCategory:",
      most_frequent_purchased_product)
```

```python
 #Calculate the average transaction amount
average_transaction_amount = df['total_amount'].mean()
print("Average Transaction Amount: ₹", average_transaction_amount)

```

```python
 #Filter transactions where the total amount is greater than ₹200
high_value_transactions = df[df['total_amount'] > 200]
print("High Value Transactions:\n", high_value_transactions)

```

```python

#Find the top 3 customers who have spent the most
customer_spending = df.groupby('CustomerID')['total_amount'].sum()
top_customers = customer_spending.nlargest(3)
print("Top 3 Customers by Spending:\n", top_customers)


```
